```python
def calculate_r_squared(y, y_pred):

    y_mean = np.mean(y)
    print('Cytof:  ',y)
    print('mean Cytof:  ',y_mean)
    # Calculate the total sum of squares (SST)
    ss_total = np.sum((y - y_mean) ** 2)
    # Calculate the sum of squares of residuals (SSR)
    ss_residual = np.sum((y - y_pred) ** 2)
    print('total sum of squares:  ',ss_total)
    print('squares of residuals:  ',ss_residual)
    # Calculate R-squared
    r_squared = 1 - (ss_residual / ss_total)
    print('R^2:  ' ,r_squared)

calculate_r_squared(WT_NK_72h[0], res1[2])
```
