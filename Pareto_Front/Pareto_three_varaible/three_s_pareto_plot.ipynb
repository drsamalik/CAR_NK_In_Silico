```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Load and sort data
df0 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_ET_10.xlsx', index_col=False)
df0.sort_values(by='alpha1', inplace=True)
df = df0

# Set up figure and axis
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
plt.rcParams['axes.linewidth'] = 5
ax.tick_params(axis='both', which='major', labelsize=18, length=6, width=5)

# Normalize color scale
norm = mcolors.Normalize(vmin=0.0, vmax=1.0)

# Diagonal reference line
x = np.arange(0, 105)
ax.plot(x, x, color='gray', lw=4, alpha=0.15)

# Scatter plot with color and size encoding
sc = ax.scatter(df['Healthy_Lysis'], df['Tumor_lysis'],
                c=df['alpha1'], s=df['alpha4'] * 80,
                cmap='coolwarm', norm=norm, label='ET=10:1')

# Colorbar with font customization
cbar = fig.colorbar(sc)
cbar.ax.tick_params(labelsize=18)  # Tick font size
#cbar.set_label('$r\alpha_1$', fontsize=20)  # Optional colorbar label

# Optional: force specific tick locations (uncomment to use)
# cbar.set_ticks([0.0, 0.5, 1.0])
# cbar.set_ticklabels(['0.0', '0.5', '1.0'])

# Axes ticks
plt.xticks([25, 50, 75], fontsize=20)
plt.yticks([25, 50, 75], fontsize=20)

# Legend
plt.legend(fontsize=12, loc='lower right', bbox_to_anchor=(1.0, 0.0))

# Axis limits
plt.xlim(20, 85)
plt.ylim(20, 85)

# Quiver plot (arrows for 'w')
plt.quiver(df['Healthy_Lysis'], df['Tumor_lysis'],
           np.zeros(len(df)), 8 * df['w'],
           angles='xy', scale_units='xy', scale=1,
           color='gray', alpha=0.35)

# Save and show plot
plt.savefig('Three_Pareto_f_ET10.png', bbox_inches='tight')
plt.show()
```
